```python
import io
df = pd.read_csv(io.BytesIO(uploaded['Capstone_E-Commerce_zscore_4Cluster.csv']))
df.info()
```

```python
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
y=df.Label
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.8)
```

```python
from sklearn.cluster import AgglomerativeClustering
model2= AgglomerativeClustering(n_clusters=4, affinity= "euclidean", linkage= "ward")
import numpy as np
import seaborn as sns; sns.set(style="ticks", color_codes=True)
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
model2.fit(X_train)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
plt.figure(figsize=(18,16))
plt.title("Agglomerative Hierarchial Clustering Dendogram")
plt.xlabel("Sample index")
plt.ylabel("Distance")
Z=linkage(X_train, "ward")
dendrogram(Z, leaf_rotation=90, p=25, color_threshold=12,
         leaf_font_size=15, truncate_mode="level")
plt.tight_layout()
```
